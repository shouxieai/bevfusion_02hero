```python
import torch
import matplotlib.pyplot as plt
import pickle
```

```python
"""
此points.pkl，是base.py 第106行没运行时，导出的points
"""
with open("points.pkl", "rb") as f:
    content = f.read()
    res = pickle.loads(content)
res = res.squeeze().to("cpu")
print(res.shape)
```

```python

"""
可视化没有经过相机内参矩阵的点
"""
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection="3d")
ax.set_xlabel('Depth')
ax.set_ylabel('width')
ax.set_zlabel('Height')

D, H, W, coor = res.shape
for d_index in range(D):
    points_dIdx = res[d_index]
    points_dIdx = points_dIdx.reshape(-1, 3)
    
    # print(points_dIdx.shape) # torch.Size([2816, 3])
    x, y, d = points_dIdx.T
    
    # print(min(x))
    # print(max(x))
    # print(min(y))
    # print(max(y))
    # print(min(d))
    # print(max(d))
    
    ax.scatter(d, x, y, marker="o", s=1) 
    ax.view_init(elev=90, azim=90)
plt.show()
fig.savefig('points_without_intrins.png')   
    
```
